```python
import requests  #Use the requests library to download web pages
from bs4 import BeautifulSoup  # Python library for pulling data out of HTML and XML files
import pandas as pd 
```

```python
airlines_urls = 'https://www.airlinequality.com/'
```

```python
response = requests.get(airlines_urls)
```

```python
page_contents = response.text
page_contents[:1000]
```

```python
with open('webpage.html', 'w', encoding='utf-8') as f:
    f.write(page_contents)
```

```python
doc = BeautifulSoup(page_contents , 'html.parser')
```

```python
doc
```

```python
airline_title_tag = doc.find_all('h2',{'class':'text_header'})
airline_title_tag
```
