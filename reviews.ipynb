```python
import requests  #Use the requests library to download web pages
from bs4 import BeautifulSoup  # Python library for pulling data out of HTML and XML files
import pandas as pd
```

```python
airlines_urls = 'https://www.airlinequality.com/review-pages/a-z-airline-reviews/'
base_url = 'https://www.airlinequality.com/'
```

```python
response = requests.get(airlines_urls)
```

```python
page_contents = response.text
page_contents[:1000]
```

```python
with open('webpage.html', 'w', encoding='utf-8') as f:
    f.write(page_contents)
```

```python
doc = BeautifulSoup(page_contents , 'html.parser')
```

```python
doc
```

```python
airline = doc.find_all('div',{'class':'a_z_col_group'})
airline

```

```python
# Assuming airline is a ResultSet object containing multiple div elements
airline_name_list = []

# Iterate over each div element in the ResultSet
for airline_div in airline:
    # Find all <ul> tags with class 'items' within the current div element
    ul_tags = airline_div.find_all('a')
    # Extend the airline_name_list with the ul_tags found in the current div element
    airline_name_list.extend(ul_tags)

# airline_name_list now contains all the <ul> tags with class 'items' found in all the div elements
```

```python
airline_name_list
```

```python
for link in airline_name_list:
       print(link['href'])
```

```python
airline_review_page_url = []
for tag in airline_name_list:
    airline_review_page_url.append(base_url + tag['href'])
print(airline_review_page_url)
```

```python
response = requests.get(airline_review_page_url[0])
```

```python
review_page_contents = response.text
review_page_contents[:1000]
```
